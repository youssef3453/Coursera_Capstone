<h1>the first question 1</h1>

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header=0)[0] 
df.head()
```

<h3>Delete rows with Not assigend Borough</h3>

```python
df=df[df["Borough"] != "Not assigned"]
df.index = [x for x in range(df.shape[0])] 
```

```python
df1 = pd.read_csv('Geospatial_Coordinates .csv') #gettin postal codes data
```

```python
ordered_postal_code = list(df["Postal Code"]) # getting a list of postal codes out of df
```

<h1>second question</h1>

```python
df1 = df1.set_index("Postal Code").reindex(ordered_postal_code).reset_index()
# changing the order of the rows in df1 to match the same order of df according to postal code column
```

```python
df1.head()
```

```python
lat_lan_cols = df1[['Latitude', 'Longitude']]
```

```python
lat_lan_cols.head()
```

```python
df2 = pd.concat([df, lat_lan_cols], axis=1)
```

```python
df2.head()
```

```python
df2_hotone = pd.get_dummies(df2)
df2_hotone.head()
```

```python
kclusters = 5

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df2_hotone)

kmeans.labels_[0:10] 
```

```python
df2_hotone.insert(0, 'Cluster Labels', kmeans.labels_)
df2_hotone.head()
```

```python
df2_hotone = pd.concat([df2_hotone, df2[["Neighbourhood"]]], axis=1)
```

```python
df2_hotone.head()
```

# third question

```python
k_means_cluster_centers = kmeans.cluster_centers_
k_means_labels = kmeans.labels_
```

```python
latitude, longitude = 43.733301, -79.372345 #coordinates of Toronto
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, name, cluster in zip(df2_hotone['Latitude'], df2_hotone['Longitude'], df2_hotone['Neighbourhood'], df2_hotone['Cluster Labels']):
    label = folium.Popup(str(name) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
